```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
angle=np.radians(np.arange(0,721))
```

```python
sine=np.sin(angle)
cosine=np.cos(angle)
```

```python
plt.plot(angle,sine,label='sine')
plt.plot(angle,cosine,label='cosine')
plt.legend()
plt.xlabel('')
plt.show()
```

```python
cars=pd.read_csv("C:\\Users\\pgkvi\\OneDrive\\Desktop\\Python Practice\\DS course\\DataSet\\usedcars.csv")
cars.head()
```

```python
plt.scatter(x='price',y='mileage',data=cars,c='purple')
plt.xlabel('car prices (USD)')
plt.ylabel('Car Mileage (MILES)')
plt.title("Scatter Plot Ex")
plt.show()
```

```python
plt.scatter(x='price',y='mileage',data=cars,c='purple',marker='*')
plt.xlabel('car prices (USD)')
plt.ylabel('Car Mileage (MILES)')
plt.title("Scatter Plot Ex")
plt.show()
```

```python
plt.scatter(x='price',y='mileage',data=cars[cars.model=='SE'],label='SE')
plt.scatter(x='price',y='mileage',data=cars[cars.model=='SES'],label='SES')
plt.scatter(x='price',y='mileage',data=cars[cars.model=='SEL'],label='SEL')
plt.legend()
plt.xlabel('car prices (USD)')
plt.ylabel('Car Mileage (MILES)')
plt.title("Scatter Plot Ex")
plt.show()
```

```python
models=['SE','SES','SEL']
for m in models:
    plt.scatter(x='price',y='mileage',data=cars[cars.model==m],label=m)
plt.legend()
plt.xlabel('car prices (USD)')
plt.ylabel('Car Mileage (MILES)')
plt.title("Scatter Plot Ex")
plt.show()
```

```python
models=[['SE','orange'],['SES','blue'],['SEL','purple']]
for m in models:
    plt.scatter(x='price',y='mileage',data=cars[cars.model==m[0]],label=m[0],c=m[1])#mention the index
plt.legend(title='car models')
plt.xlabel('car prices (USD)')
plt.ylabel('Car Mileage (MILES)')
plt.title("Scatter Plot Ex")
plt.show()
```

```python
models=[['SE','orange','^'],['SES','blue','*'],['SEL','purple','p']]
for m in models:
    plt.scatter(x='price',y='mileage',data=cars[cars.model==m[0]],label=m[0],c=m[1],marker=m[2])#mention the index
plt.legend(title='car models',ncol=2,shadow=True)
plt.xlabel('car prices (USD)')
plt.ylabel('Car Mileage (MILES)')
plt.title("Scatter Plot Ex")
plt.show()
```
