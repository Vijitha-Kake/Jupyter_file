```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier,plot_tree,export_text
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay
```

```python
fruits=pd.read_csv("C:\\Users\\pgkvi\\OneDrive\\Desktop\\Python Practice\\DS course\\DataSet\\fruits.csv")
fruits.head()
```

```python
#aft EDa we took the features and y label 
x=fruits[['sphericity','weight']]
y=fruits['label']
```

```python
sns.scatterplot(x='sphericity',y='weight',data=fruits,hue='label')
plt.show()
```

```python
#(splitting the data)training and testing the data for evalution 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
y_train.shape,y_test.shape,x_train.shape,x_test.shape
```

```python
x_train.index,x_train.index
```

```python
#modeltraining fitting
model=DecisionTreeClassifier(criterion='entropy')
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
plt.figure(figsize=[6,8])
plot_tree(decision_tree=model,feature_names=['sphericity','weight'],class_names=['Apple','Orange'])
plt.show()
```

```python
plt.figure(figsize=[6,8])
plot_tree(decision_tree=model,feature_names=model.feature_names_in_,class_names=model.classes_)
plt.show()
```

```python
sns.scatterplot(x='sphericity',y='weight',data=fruits,hue='label')
sns.scatterplot(x='sphericity',y='weight',data=x_test,color='blue',marker='*',s=100,label='Test Samples')
plt.plot([0.7,0.95],[138,138],c='purple')
plt.plot([0.7,0.95],[147,147],c='purple')
plt.plot([0.796,0.796],[138,147],c='purple')
plt.legend()
plt.show()
```

```python
y_predict = model.predict(x_test)
y_predict
```

```python
score=(y_predict==y_test).sum()/len(y_test)
score
```

```python
accuracy_score(y_test,y_predict)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_predict)
plt.show()
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_predict)
plt.show()
```

```python
plt.scatter(x='sphericity',y='weight',data=x_train[y_train=='Apple'],c='red',label='Apple train samples')
plt.scatter(x='sphericity',y='weight',data=x_train[y_train=='Orange'],c='orange',label='Orange train samples')
sns.scatterplot(x='sphericity',y='weight',data=x_test[y_test==y_predict],color='blue',marker='*',s=120,label='Test Correct Samples')
sns.scatterplot(x='sphericity',y='weight',data=x_test[y_test!=y_predict],color='green',marker='*',s=120,label='Test Wrong Samples')
plt.legend()
plt.show()
```
