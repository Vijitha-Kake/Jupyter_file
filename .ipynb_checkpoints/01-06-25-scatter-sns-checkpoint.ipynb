```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
iris.species.value_counts()
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
#matplotlib
plt.scatter(x='sepal_length',y='petal_length',data=iris[iris.species=='setosa'],label='setosa')
plt.scatter(x='sepal_length',y='petal_length',data=iris[iris.species=='versicolor'],label='versicolor')
plt.scatter(x='sepal_length',y='petal_length',data=iris[iris.species=='virginica'],label='virginica') 
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('IRIS DATASET')
plt.legend(title='species')
plt.show()
```

```python
plt.scatter(x='sepal_length',y='petal_length',data=iris[iris.species=='setosa'],c='blue',s=20,label='setosa')
plt.scatter(x='sepal_length',y='petal_length',data=iris[iris.species=='versicolor'],c='blue',s=40,label='versicolor')
plt.scatter(x='sepal_length',y='petal_length',data=iris[iris.species=='virginica'],c='blue',s=60,label='virginica') 
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('IRIS DATASET')
plt.legend(title='species')
plt.show()
```

```python
species=['setosa','versicolor','virginica']
for i in species:
    plt.scatter(x='sepal_length',y='petal_length',data=iris[iris.species==i],label=i)
plt.legend()
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='species')#for different color we use hue
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,style='species')#for different marker we use style
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,size='species')#for different size we use size
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,size='species',hue='species')#for different marker we use style
plt.show()
```

```python
sns.countplot(x='day',data=tips,hue='time')
plt.show()
```

```python
df1=tips.groupby('day').max(numeric_only=True)['tip'].to_frame().reset_index()
df1
```

```python
sns.barplot(x='day',y='tip',data=df1)
plt.ylabel('Max Tip')
plt.show()
```

```python
df2=tips.groupby('day').mean(numeric_only=True)['tip'].to_frame().reset_index()
df2
```

```python
sns.barplot(x='day',y='tip',data=df2)
plt.ylabel('Mean Tip Values')
plt.show()
```
