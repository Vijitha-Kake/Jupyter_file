```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
airlines=pd.read_csv("C:\\Users\\pgkvi\\OneDrive\\Desktop\\Python Practice\\DS course\\DataSet\\Invistico_Airline.csv")
airlines.head()
```

```python
plt.plot('Flight Distance',data=airlines[airlines.satisfaction=='satisfied'])
plt.show()
```

```python
df1=airlines.satisfaction.value_counts().to_frame().reset_index()
df1
```

```python
plt.bar(x='index',height='satisfaction',data=df1)
for i in range(len(df1)):
    plt.text(x=i,y=df1['satisfaction'][i]+400,s=df1['satisfaction'][i],ha='center')
plt.show()
```

```python
df_male=airlines[airlines.Gender=='Male'].satisfaction.value_counts(sort=False).to_frame().reset_index()
df_male
```

```python
df_female=airlines[airlines.Gender=='Female'].satisfaction.value_counts().to_frame().reset_index()
df_female
```

```python
plt.bar(x='index',height='satisfaction',data=df_male,color='blue',width=0.4,align='edge')
for i in range(len(df_male)):
    plt.text(x=i+0.2,y=df_male['satisfaction'][i]+400,s=df_male['satisfaction'][i],ha='center')
plt.bar(x='index',height='satisfaction',data=df_female,color='pink',align='edge',width=-0.4)
for i in range(len(df_female)):
    plt.text(x=i-0.2,y=df_female['satisfaction'][i]+400,s=df_female['satisfaction'][i],ha='center')
plt.show()
```

```python
plt.bar(x='index',height='satisfaction',data=df_male)
#for i in range(len(df_male)):
#    plt.text(x=i,y=df_male['satisfaction'][i]+400,s=df_male['satisfaction'][i],ha='center')
plt.show()
```

```python
plt.bar(x='index',height='satisfaction',data=df_male,color='blue',bottom=df_female['satisfaction'],label='Male')
plt.bar(x='index',height='satisfaction',data=df_female,color='pink',label='Female')
for i in range(2):
    plt.text(x=i,y=df_female['satisfaction'][i]/2,s=df_female['satisfaction'][i],ha='center')
    plt.text(x=i,y=df_female['satisfaction'][i]+df_male['satisfaction'][i]/2,s=df_male['satisfaction'][i],ha='center')
    plt.text(x=i,y=df_female['satisfaction'][i]+df_male['satisfaction'][i]+400,
             s=df_female['satisfaction'][i]+df_male['satisfaction'][i],ha='center')
plt.legend(title='Gender')
plt.show()
```

```python
plt.barh(y='index',width='satisfaction',data=df_male,color='blue',left=df_female['satisfaction'],label='Male')
plt.barh(y='index',width='satisfaction',data=df_female,color='pink',label='Female')
for i in range(2):
    plt.text(y=i,x=df_female['satisfaction'][i]/2,s=df_female['satisfaction'][i],va='center')
    plt.text(y=i,x=df_female['satisfaction'][i]+df_male['satisfaction'][i]/2,s=df_male['satisfaction'][i],va='center')
    plt.text(y=i,x=df_female['satisfaction'][i]+df_male['satisfaction'][i]+400,
             s=df_female['satisfaction'][i]+df_male['satisfaction'][i],va='center')
plt.legend(title='Gender')
plt.show()
```
