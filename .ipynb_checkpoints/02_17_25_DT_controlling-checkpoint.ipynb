```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier,plot_tree,export_text
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay
```

```python
fruits=pd.read_csv("C:\\Users\\pgkvi\\OneDrive\\Desktop\\Python Practice\\DS course\\DataSet\\fruits.csv")
fruits.head()
```

```python
x=fruits[['sphericity','weight','color']]
y=fruits['label']
x.tail()
```

```python
sns.scatterplot(x='sphericity',y='weight',data=fruits,hue='label')
plt.show()
```

```python
for column in x.columns:
    if x[column].dtype==type(object):
        le=LabelEncoder()
        x[column]=le.fit_transform(x[column])
```

```python
x.tail()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10,shuffle=True)
x_train.index,x_test.index,y_train.index,y_test.index
```

```python
model=DecisionTreeClassifier(criterion='entropy')
model.fit(x_train,y_train)
```

```python
x.dtypes
```

```python
model.score(x_test,y_test)
```

```python
plt.figure(figsize=[6,8])
plot_tree(decision_tree=model,feature_names=model.feature_names_in_,class_names=model.classes_)
plt.show()
```

```python
model.get_depth(),model.get_n_leaves()
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
accuracy_score(y_test,y_predict)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_predict)
plt.show()
```

```python
plt.scatter(x='sphericity',y='weight',data=x_train[y_train=='Apple'],c='red',label='Apple Samples')
plt.scatter(x='sphericity',y='weight',data=x_train[y_train=='Orange'],c='green',label='Orange Samples')
plt.scatter(x='sphericity',y='weight',data=x_test[y_test==y_predict],c='blue',marker='*',label='Correctly Samples')
plt.scatter(x='sphericity',y='weight',data=x_test[y_test!=y_predict],c='orange',marker='*',label='Wrongly Samples')
plt.legend()
plt.show()
```

```python
#if we get too mnay no of depth we can control decision tree depth by giving max_depth
model=DecisionTreeClassifier(criterion='entropy',max_depth=2)
model.fit(x_train,y_train)
model.score(x_train,y_train),model.score(x_test,y_test)
```

```python
plt.figure(figsize=[6,8])
plot_tree(model,feature_names=model.feature_names_in_,class_names=model.classes_)
plt.show()
```

```python
y_predict=model.predict(x_test)
y_predict
#accuracy_score(y_test,y_predict)
```
