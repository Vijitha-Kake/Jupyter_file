```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars=pd.read_csv("C:\\Users\\pgkvi\\OneDrive\\Desktop\\Python Practice\\DS course\\DataSet\\usedcars.csv")
cars.head()
```

```python
plt.boxplot(x='price',data=cars)
plt.title('Price Data')
plt.show()
plt.boxplot(x='price',data=cars,widths=0.3)
plt.title('Price Data')
plt.show()
```

```python
plt.boxplot(x='price',data=cars,sym='D')
plt.title('Price Data')
plt.show()
plt.boxplot(x='price',data=cars,sym='D',vert=False)
plt.title('Price Data')
plt.show()
```

```python
plt.boxplot(x='price',data=cars,showfliers=False)
plt.title('Price Data')
plt.show()
```

```python
plt.boxplot(x='price',data=cars,whis=1)#default value is 1.5times of IQR value
plt.title('Price Data')
plt.show()
plt.boxplot(x='price',data=cars,whis=0)
plt.title('Price Data')
plt.show()
```

```python
plt.boxplot(x='price',data=cars[cars.model=='SE'],positions=[0],widths=0.4)
plt.boxplot(x='price',data=cars[cars.model=='SES'],positions=[1],widths=0.4)
plt.boxplot(x='price',data=cars[cars.model=='SEL'],positions=[2],widths=0.4)
plt.xlabel('Car Model')  # Ensure this is not overridden elsewhere
plt.ylabel('Price Value')
plt.title('Box Plot for Car Model')
plt.show()
```
