```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cars=pd.read_csv("C:\\Users\\pgkvi\\OneDrive\\Desktop\\Python Practice\\DS course\\DataSet\\usedcars.csv")
cars.head()
```

```python
plt.boxplot(x='price',data=cars,widths=0.1)
plt.show()
```

```python
plt.boxplot(x='price',data=cars,showfliers=True,whis=0.5,notch=True)
plt.show()
```

```python
plt.boxplot(x='price',data=cars,showfliers=False,whis=0.5,notch=True)
plt.show()
```

```python
plt.boxplot(x='price',data=cars[cars.model=='SE'],widths=0.4,positions=[0])
plt.boxplot(x='price',data=cars[cars.model=='SES'],widths=0.4,positions=[1])
plt.boxplot(x='price',data=cars[cars.model=='SEL'],widths=0.4,positions=[2])
plt.xlabel('car models')
plt.ylabel('price')
plt.title('Box Plot for Cars Model')
plt.xticks(ticks=[0,1,2],labels=['SE','SES','SEL'])
plt.show()
```

```python
for m in ['SE','SES','SEL']:
    plt.boxplot(x='price',data=cars[cars.model==m],widths=0.4,positions=[['SE','SES','SEL'].index(m)])
plt.xlabel('car models')
plt.ylabel('price')
plt.title('Box Plot for Cars Model')
plt.xticks(ticks=[0,1,2],labels=['SE','SES','SEL'])
plt.show()
```
