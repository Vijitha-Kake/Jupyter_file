```python
import pandas as pd
```

```python
cars=pd.read_csv("C:\\Users\\pgkvi\\OneDrive\\Desktop\\Python Practice\\DS course\\DataSet\\usedcars.csv")
cars.head()
```

```python
cars[['model','color','price']].groupby(['model','color']).mean()#or
```

```python
cars.groupby(['model','color']).mean()[['price']]
```

```python
cars.pivot_table(index='model',columns='color',values='price',aggfunc='mean')
```

```python
cars.pivot_table(index='model',columns='color',values='price',aggfunc='count')
```

```python
cars.pivot_table(index='model',columns='color',values='price',aggfunc='count',fill_value=0)#or
```

```python
cars.pivot_table(index='model',columns='color',values='price',aggfunc='count').fillna(value=0)
```

```python
#to get the row total we use margin as true
cars.pivot_table(index='model',columns='color',values='price',aggfunc='count',fill_value=0,
                 margins=True,margins_name='Total_count')
```
