```python
#matplotlib
plt.scatter(x='sepal_length',y='petal_length',data=iris[iris.species=='setosa'],label='setosa')
plt.scatter(x='sepal_length',y='petal_length',data=iris[iris.species=='versicolor'],label='versicolor')
plt.scatter(x='sepal_length',y='petal_length',data=iris[iris.species=='virginica'],label='virginica') 
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('IRIS DATASET')
plt.legend(title='species')
plt.show()
```

```python
species=['setosa','versicolor','virginica']
for i in species:
    plt.scatter(x='sepal_length',y='petal_length',data=iris[iris.species==i],label=i)
plt.legend()
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='species')
plt.show()
```
