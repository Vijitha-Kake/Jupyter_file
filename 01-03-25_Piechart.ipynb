```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars=pd.read_csv("C:\\Users\\pgkvi\\OneDrive\\Desktop\\Python Practice\\DS course\\DataSet\\usedcars.csv")
cars.head()
```

```python
df1=cars.model.value_counts().to_frame().reset_index()
df1
```

```python
plt.pie(x='model',data=df1,labels='index',colors=['orange','green','purple'])
plt.pie(x='model',data=df1,labels='model',radius=0.4,colors=['orange','green','purple'])
plt.show()
```

```python
plt.pie(x='model',data=df1,labels='index',shadow=True,explode=[0,0,0.12])
#plt.pie(x='model',data=df1,labels='model',radius=0.4,colors=['orange','green','purple'])
plt.show()
```

```python
plt.hist(x='price',data=cars)
plt.show()
```

```python
plt.hist(x='price',data=cars[cars.model=='SE'],bins=15,histtype='step',label='SE')
plt.hist(x='price',data=cars[cars.model=='SES'],bins=15,histtype='step',label='SES')
plt.hist(x='price',data=cars[cars.model=='SEL'],bins=15,histtype='step',label='SEL')
plt.legend()
plt.show()
```
