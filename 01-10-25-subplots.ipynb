```python
plt.figure(figsize=[12,8])
plt.subplot(2,2,1)#witha 2unit axis 1
plt.scatter(x='price',y='mileage',data=cars)
plt.subplot(2,2,2)
plt.boxplot(x='price',data=cars)
plt.subplot(2,2,3)
plt.hist(x='price',data=cars)
plt.subplot(2,2,4)
plt.plot('price','mileage',data=cars)
plt.show()
```
