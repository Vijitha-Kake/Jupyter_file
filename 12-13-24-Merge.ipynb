```python
import pandas as pd
```

```python
emp=pd.DataFrame({'emp.no.':[1001,1002,1003,1004,1005,1006,1007],
                          'emp.name':['abc','def','ghi','jkl','mnl','opq','rst'],
                          'sal':[2000,2100,2200,2300,2400,2500,2600],
                          'deptno.':[20,30,10,30,10,20,50]})
emp
```

```python
dept=pd.DataFrame({'deptno.':[10,20,30,40],
                  'Dname':['accounts','sales','research','operations'],
                  'location':['Hyd','Chennai','Mumbai','Bengaluru']})
dept
```

```python
pd.merge(left=emp,right=dept,on='deptno.',how='inner')
```

```python
pd.merge(left=emp,right=dept,on='deptno.',how='outer')
```

```python
print(pd.merge(left=emp,right=dept,on='deptno.',how='left'))
pd.merge(left=emp,right=dept,on='deptno.',how='right')
```

```python
emp=pd.DataFrame({'emp.no.':[1001,1002,1003,1004,1005,1006,1007],
                          'emp.name':['abc','def','ghi','jkl','mnl','opq','rst'],
                          'sal':[2000,2100,2200,2300,2400,2500,2600],
                          'deptno.':[20,30,10,30,10,20,50]})
emp
```

```python
dept=pd.DataFrame({'dno.':[10,20,30,40],
                  'Dname':['accounts','sales','research','operations'],
                  'location':['Hyd','Chennai','Mumbai','Bengaluru']})
dept
```

```python
pd.merge(left=emp,right=dept,left_on='deptno.',right_on='dno.',how='inner') #left_on nd right_on based on the columns
```

```python
#merging the table based on index
emp=pd.DataFrame({'emp.no.':[1001,1002,1003,1004,1005,1006,1007],
                          'emp.name':['abc','def','ghi','jkl','mnl','opq','rst'],
                          'sal':[2000,2100,2200,2300,2400,2500,2600]},
                          index=[20,30,10,30,10,20,50])
emp
```

```python
dept=pd.DataFrame({'Dname':['accounts','sales','research','operations'],
                  'location':['Hyd','Chennai','Mumbai','Bengaluru']},
                 index=[10,20,30,40])
dept
```

```python
pd.merge(left=emp,right=dept,left_index=True,right_index=True,how='inner')
```

```python
pd.merge(left=emp,right=dept,left_index=True,right_index=True,how='left')
```
