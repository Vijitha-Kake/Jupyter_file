```python
import pandas as pd
```

```python
students=pd.DataFrame({'sname':['viji','naz','subbu','vijay'],
                     'maths':[99,96,98,94],
                      'physics':[96,97,94,97],
                      'chemistry':[98,96,97,95]})
students
```

```python
stu=students.melt(id_vars='sname',var_name='subject',value_name='marks')
stu
```

```python
stu.groupby('subject').describe()
```

```python
stu.groupby('sname').describe()
```

```python
stu=students.melt(id_vars='sname',value_vars=['physics','maths'],var_name='subject',value_name='marks')
stu
```
